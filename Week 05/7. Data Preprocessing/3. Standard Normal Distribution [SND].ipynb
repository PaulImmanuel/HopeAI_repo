```python
import pandas as pd
dataset = pd.read_csv("PrePlacement.csv")
```

```python
dataset
```

```python
def stdNBgraph(dataset):
     from sklearn.preprocessing import StandardScaler
     sc = StandardScaler()
     sc.fit(dataset)
     return sc.transform(dataset)
```

```python
SNB = stdNBgraph(dataset['salary'].to_frame())
```

```python
SNB
```

```python
import seaborn as sns
sns.displot(SNB,kde=True)
```

```python
import numpy as np
newSal = np.log(SNB + 1)
```

```python
sns.displot(newSal,kde=True)
```
