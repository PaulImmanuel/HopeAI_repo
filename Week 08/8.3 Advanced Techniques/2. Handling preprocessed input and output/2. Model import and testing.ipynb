```python
import pickle
import warnings

warnings.filterwarnings("ignore")

loaded_model = pickle.load(
    open("pre-processed_io.sav","rb")
)

loaded_scx = pickle.load(
    open("pre-processed_io_scx.sav","rb")
)

loaded_scy = pickle.load(
    open("pre-processed_io_scy.sav","rb")
)
```

```python
test_input = loaded_scx.transform([[
    120000,
    110000,
    300000,
    0,
    1
]])
```

```python
scaled_output = loaded_model.predict(
    test_input
)

final_output = loaded_scy.inverse_transform(
    scaled_output.reshape(-1,1)
)

print(final_output)
```
