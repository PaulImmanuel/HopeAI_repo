```python
import pandas as pd

# Load dataset
data = pd.read_csv("50_Startups.csv")

# Convert categorical values into numerical format
data = pd.get_dummies(data, drop_first=True)

# Selecting input and output columns
X = data[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]

y = data[['Profit']]

# Splitting dataset into training and testing data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0
)

# Import and train Multiple Linear Regression model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)

# Model coefficients
coefficients = model.coef_

# Intercept value
intercept = model.intercept_

# Predict test data
predictions = model.predict(x_test)

# Calculate accuracy using R2 score
from sklearn.metrics import r2_score

score = r2_score(y_test, predictions)

# Save trained model
import pickle

model_file = "finalized_MLR_model.sav"
pickle.dump(model, open(model_file, "wb"))

# Load saved model
saved_model = pickle.load(open(model_file, "rb"))

# Example prediction
# Input format:
# [R&D Spend, Administration, Marketing Spend, State_Florida, State_New York]

output = saved_model.predict([[1234, 345, 4565, 1, 0]])

print(output)
```
