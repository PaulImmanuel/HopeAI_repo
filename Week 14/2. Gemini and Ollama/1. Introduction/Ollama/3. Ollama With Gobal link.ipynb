```python
import ollama
import gradio as gr

# System message (for context)
system_msg = "You are a Sports equipment seller. You are to suggest relevant products based on user's sport."

# Chatbot function
def CustomLlama3(user_input):
    # Format the conversation context
    messages = [
        {"role": "system", "content": system_msg},
        {"role": "user", "content": user_input}
    ]

    # Generate response using Ollama's Llama 3 model
    response = ollama.chat(model="llama3", messages=messages)

    # Extract and return response text
    return response['message']['content']

# Gradio Interface
demo = gr.Interface(fn=CustomLlama3, 
                    inputs="text", 
                    outputs="text", 
                    title="My Sports Assistant")

# Launch the app
demo.launch(share=True)
```
