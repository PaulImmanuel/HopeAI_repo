```python
import google.generativeai as genai
import gradio as gr

# Configure API key
genai.configure(api_key="zzzzzzzzzzzzzzzzzzzzzzzzzzzz")  # Replace with your actual Gemini API key

# Load the model
model = genai.GenerativeModel(model_name="gemini-2.5-flash")

# System message (used only for context)
system_msg = "You are a gym training assistant. you are to suggest equipments to new users."

# Chatbot function
def CustomGemini(user_input):
    response = model.generate_content(system_msg + "\n" + user_input)
    return response.text  # Extract and return response

# Gradio Interface
demo = gr.Interface(fn=CustomGemini, 
                    inputs="text", 
                    outputs="text", 
                    title="My Gym training Assistant")

# Launch the app
demo.launch(share=True)
```
