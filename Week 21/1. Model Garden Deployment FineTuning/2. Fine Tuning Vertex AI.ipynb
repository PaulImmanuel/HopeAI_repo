```python
from google.cloud import aiplatform

PROJECT_ID = "automlvertexai"
REGION = "us-central1"
ENDPOINT_ID = "1316441423645507584"  # 🔴 replace this

# Initialize
aiplatform.init(project=PROJECT_ID, location=REGION)

# Load endpoint
endpoint = aiplatform.Endpoint(ENDPOINT_ID)

# Send prompt
response = endpoint.predict(
    instances=[
        {
            "prompt": "tell about Hope AI compny",
            "max_tokens": 100,
            "temperature": 0.7
        }
    ]
)

print(response)
```
