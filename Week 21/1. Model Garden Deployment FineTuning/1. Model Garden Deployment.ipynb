```python
from google.cloud import aiplatform

PROJECT_ID = "agentplatformmodel"
REGION = "us-central1"
ENDPOINT_ID = "mg-endpoint-16a5fcae-f87c-4028-a5a1-63fb8e92d12a"  # 🔴 replace this

# Initialize
aiplatform.init(project=PROJECT_ID, location=REGION)

# Load endpoint
endpoint = aiplatform.Endpoint(ENDPOINT_ID)

# Send prompt
response = endpoint.predict(
    instances=[
        {
            "prompt": "Explain Artificial Intelligence in simple terms",
            "max_tokens": 100,
            "temperature": 0.7
        }
    ]
)

print(response)
```
