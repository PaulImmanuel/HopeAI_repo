```python
import pickle
import pandas as pd

model = pickle.load(open("AttritionModel.sav","rb"))
selected_features = pickle.load(open("SelectedFeatures.sav","rb"))
encoder_columns = pickle.load(open("encoder_columns.sav","rb"))
```

```python
dataset = pd.read_csv("HR-Employee-Attrition.csv")
dataset.head()
```

### Inputs

```python
age = int(input("Enter Age : "))
gender = input("Gender (Male/Female): ").title()
department = input("Department (Sales / Research & Development / Human Resources): ")
job_role = input("Job Role: ")
education_field = input("Education Field: ")
marital_status = input("Marital Status (Single / Married / Divorced): ")
overtime = input("OverTime (Yes / No): ").title()
```

```python
print("\n======== Employee Details =========")
print("Age             :", age)
print("Gender          :", gender)
print("Department      :", department)
print("Job Role        :", job_role)
print("Education Field :", education_field)
print("Marital Status  :", marital_status)
print("OverTime        :", overtime)
print("=====================================")
```

```python
user_data = {"Age":age}

for col in encoder_columns:
    if col not in user_data:
        user_data[col] = 0
```

### input encoding

```python
if gender == "Male":
    user_data["Gender_Male"] = 1


if department == "Research & Development":
    user_data["Department_Research & Development"] = 1
elif department == "Sales":
    user_data["Department_Sales"] = 1

#######################
if education_field == "Life Sciences":
    user_data["EducationField_Life Sciences"] = 1
elif education_field == "Marketing":
    user_data["EducationField_Marketing"] = 1
elif education_field == "Medical":
    user_data["EducationField_Medical"] = 1
elif education_field == "Other":
    user_data["EducationField_Other"] = 1
elif education_field == "Technical Degree":
    user_data["EducationField_Technical Degree"] = 1

##################

if education_field == "Life Sciences":
    user_data["EducationField_Life Sciences"] = 1
elif education_field == "Marketing":
    user_data["EducationField_Marketing"] = 1
elif education_field == "Medical":
    user_data["EducationField_Medical"] = 1
elif education_field == "Other":
    user_data["EducationField_Other"] = 1
elif education_field == "Technical Degree":
    user_data["EducationField_Technical Degree"] = 1

###########################

if marital_status == "Married":
    user_data["MaritalStatus_Married"] = 1
elif marital_status == "Single":
    user_data["MaritalStatus_Single"] = 1

############################
if overtime == "Yes":
    user_data["OverTime_Yes"] = 1
```

### Dataset Conversion

```python
input_df = pd.DataFrame([user_data])
input_df = input_df[encoder_columns]
```

```python
input_df = input_df[selected_features]

print("\nSelected Features Used:")
print(selected_features)
```

### Prediction

```python
prediction = model.predict(input_df)

if prediction[0] == 1:
    print("Employee is likely to leave the company.")
else:
    print("Employee is likely to stay in the company.")
```
