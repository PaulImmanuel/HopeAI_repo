```python
""" 
Hierarchical Clustering builds a hierarchy of clusters by either merging smaller clusters or splitting larger ones.

Agglomerative Clustering (Bottom-Up):
This approach begins with each data point treated as an individual cluster. 
The algorithm then progressively combines the most similar clusters until the desired clustering structure is achieved.

Divisive Clustering (Top-Down):
In this method, all data points initially belong to a single cluster. 
The algorithm repeatedly divides clusters into smaller groups based on their differences, creating a hierarchical structure."""

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

dataset = pd.read_csv('Mall_Customers.csv')
dataset

numeric_data = dataset[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

agglo = AgglomerativeClustering(n_clusters=5)
dataset['Cluster'] = agglo.fit_predict(numeric_data)

plt.figure(figsize=(9, 5))
plt.scatter(
    dataset['Annual Income (k$)'],
    dataset['Spending Score (1-100)'],
    c=dataset['Cluster'],
    cmap='tab10',
    s=50
)
plt.title('Hierarchical Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```
