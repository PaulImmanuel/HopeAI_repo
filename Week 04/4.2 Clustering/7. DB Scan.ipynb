```python
""" 
DBSCAN is a density-based clustering algorithm that forms clusters by connecting points located close to one another. 
Points that do not belong to any dense region are treated as noise or outliers.

Key parameters:
eps: Defines the neighborhood radius around a point.
min_samples: Specifies the minimum number of points required within that radius for a region to be considered dense.

The algorithm evaluates each point by looking at the number of neighboring points within the eps distance. 
If a point has at least min_samples neighbors, it becomes part of a cluster; otherwise, 
it may be labeled as a border point or noise, depending on its surroundings. """

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import DBSCAN

dataset = pd.read_csv('Mall_Customers.csv')
dataset

numeric_data = dataset[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

dbscan = DBSCAN(eps=10, min_samples=5)
dataset['Cluster'] = dbscan.fit_predict(numeric_data)

plt.figure(figsize=(10, 6))
plt.scatter(
    dataset['Annual Income (k$)'],
    dataset['Spending Score (1-100)'],
    c=dataset['Cluster'],
    cmap='tab10',
    s=50
)
plt.title('DBSCAN Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```
