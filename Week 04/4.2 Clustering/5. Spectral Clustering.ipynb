```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import SpectralClustering

dataset = pd.read_csv('Mall_Customers.csv')
dataset

numeric_data = dataset[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

spectral = SpectralClustering(n_clusters=5, assign_labels='kmeans', random_state=0)
dataset['Cluster'] = spectral.fit_predict(numeric_data)

plt.figure(figsize=(9, 5))
plt.scatter(
    dataset['Annual Income (k$)'],
    dataset['Spending Score (1-100)'],
    c=dataset['Cluster'],
    cmap='tab10',
    s=50
)
plt.title('Spectral Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```
