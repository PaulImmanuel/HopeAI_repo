```python
""" 
OPTICS CLUSTERING:
is a density-based clustering technique designed to identify clusters of varying densities within a dataset.

Similar to DBSCAN, OPTICS groups points that are densely packed together while treating points in sparse regions as 
potential noise or outliers.

Instead of directly assigning points to clusters, OPTICS generates an ordered sequence of data points that 
reflects the underlying density-based structure of the dataset.

The algorithm produces a reachability plot, which helps visualize changes in density across the data. 
This plot can be used to identify meaningful clusters and understand the overall clustering pattern.
"""

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import OPTICS

dataset = pd.read_csv('Mall_Customers.csv')
dataset

numeric_data = dataset[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

optics = OPTICS(min_samples=5, xi=0.05, min_cluster_size=0.1)
dataset['Cluster'] = optics.fit_predict(numeric_data)

plt.figure(figsize=(9, 5))
plt.scatter(
    dataset['Annual Income (k$)'],
    dataset['Spending Score (1-100)'],
    c=dataset['Cluster'],
    cmap='tab10',
    s=50
)
plt.title('OPTICS Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```
