```python
"""
BIRCH CLUSTERING:

Builds a compact CF (Clustering Feature) tree by incrementally processing data points.
Organizes data into a hierarchical structure, making it efficient for large datasets.
Stores data as summarized cluster representations rather than individual points.
Performs the final clustering step using the information contained in the leaf nodes of the CF tree.
"""

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import Birch

dataset = pd.read_csv('Mall_Customers.csv')
dataset

numeric_data = dataset[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

birch = Birch(n_clusters=5)
dataset['Cluster'] = birch.fit_predict(numeric_data)

plt.figure(figsize=(8, 5))
plt.scatter(
    dataset['Annual Income (k$)'],
    dataset['Spending Score (1-100)'],
    c=dataset['Cluster'],
    cmap='tab10',
    s=50
)
plt.title('BIRCH Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```
